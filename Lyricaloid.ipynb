```python
from konlpy.tag import Okt
okt = Okt()
```

```python
import pandas as pd
Lyrics = pd.read_csv('J:/data/lyric.csv', encoding='cp949')
```

```python
Lyrics
```

```python
words_list = list()
for i in Lyrics.index:
    words = okt.nouns(Lyrics.loc[i,'Lyrics'])
    words_list.extend(words)
words_list
```

```python
print(len(words_list))
```

```python
import re
words_list_2 = []
for words in words_list:
    words_list_2.append(re.sub('[a-zA-Z0-9]','',words))

# words_list = list(filter(None, words_list))
print(words_list_2)
```

```python
words_list_3 = list(filter(None, words_list_2))
print(words_list_3)
```

```python
print(len(words_list_3))
```

```python
words_list_4 = []
for words in words_list_3:
    if words not in words_list_4:
        words_list_4.append(words)
        
print(words_list_4)
```

```python
print(len(words_list_4))
```

```python
words_data = pd.DataFrame(words_list_4, columns=['Words'])
# ['Words','Meaningless','Activity_Score','Emotional_Score']
words_data
```

```python
words_data.to_csv('J:\data\words_data.csv',index=False, encoding = 'cp949')
```
