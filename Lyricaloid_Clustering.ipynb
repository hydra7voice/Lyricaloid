```python
import pandas as pd
word_data = pd.read_csv('J:/data/words_data.csv', encoding='cp949')
word_data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

ref_data = word_data[['Activity_Score','Emotional_Score']]
model = KMeans(n_clusters=5)
predict = pd.DataFrame(model.fit_predict(ref_data))
predict.columns=['Cluster']

cluster_value = pd.concat([word_data,predict],axis=1)
cluster_value
```

```python
sns.pairplot(cluster_value,hue='Cluster')
plt.show()
```

```python
def cluster_n_count(index):
    cluster = cluster_value[cluster_value['Cluster'] == index]
    return cluster

cluster_n_count(0)
```

```python
# input appointment


# ordinary = cluster_n_count(0)
# neg_static = cluster_n_count(1)
# pos_dynamic = cluster_n_count(2)
# neg_dynamic = cluster_n_count(3)
# pos_static = cluster_n_count(4)

from random import randint
def input_func(input_keyword,input_emotion,cluster,index_number):
#     cluster_value[cluster_value['words'] == input_keyword]

    index_dup = []
    word = ''
    
    for i in index_dup:
        while i == index_number:
            index_number = randint(0,len(cluster)-1)

    index_dup.append(index_number)
    word = cluster.iloc[index_number].loc['Words']
    index_dup = []
    return word
```

```python
# input_emotion 범위 : 0,1,2,3,4
# 평서적 = 0
# 부정-정적 = 1
# 긍정-동적 = 2
# 부정-동적 = 3
# 긍정-정적 = 4

# input_tension,input_activity 범위 : -5 ~ 5

def score_check(input_emotion,input_tension, input_activity):
    cluster = cluster_n_count(input_emotion)
    index_number = randint(0,len(cluster)-1)
    keywords = []
    words_tension_score = 0
    words_emotional_score = 0
    
#     adjective_efficiency 계수 결정
    adjective_efficiency = ( cluster.iloc[index_number].loc['Adjective_Efficiency'] * 0.1 ) + 1
#     단어별 점수
    tension_score = cluster.iloc[index_number].loc['Activity_Score'] * adjective_efficiency
    emotional_score = cluster.iloc[index_number].loc['Emotional_Score'] * adjective_efficiency
#     목표 점수
    target_tension_score = input_tension * 5
    target_emotion_score = input_activity * 5
#     종합 점수 계산
    words_tension_score = words_tension_score + tension_score
    words_emotional_score = words_emotional_score + emotional_score
    
    for i in range(5):
        word = input_func(4,'보름달',cluster,index_number)
        if words_tension_score < (target_tension_score-4) or words_tension_score < (target_tension_score+4):
            if words_emotional_score < (target_emotion_score-4) or words_emotional_score < (target_emotion_score+4):
                keywords.append(word)
            else:
                words_emotional_score = words_emotional_score - emotional_score
                break
        else:
            words_tension_score = words_tension_score - tension_score
            break
    return keywords
      
#     tension_score_check(input_tension,words_tension_score,word)
#     activity_score_check(input_activity,words_activity_score,word)
```
